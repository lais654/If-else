```python
idade = int(input("Digite sua idade:"))

if idade < 16:
  print("Você é menor de idade")
elif idade >= 16 and idade < 18:
   print("Seu voto é opcional")
elif idade >=18:
   print("Seu voto é obrigatório")
else:
  print("Digite um valor compatível")
```

```python
pagamento = input("Selecione uma forma de pagamento:")
if pagamento == "cartão" or pagamento == "cartao":
  print("Processando pagamento no cartão")

elif pagamento == "boleto":
  print("Processando pagamento no boleto")
else:
  print ("Forma de pagamento não reconhecida")
```
